# <span style='background:yellow'>3 clusters

```python
df=pd.read_csv('C:/Users/zhaleh/PhD_Projects/Divergent Paths of Beat Perception_3 Modalities/results/PreprocessedResults_MusicAdded.csv')
df = df.iloc[:141, :]
df2=df[['TactSD','TactIrreg','TactReg']]
df2.head()
```

```python
model2=KMeans(n_clusters=3, init='k-means++',max_iter=1000, random_state=27)
model2.fit(df2)
```

```python
df2['Cluster']=model2.fit_predict(df2)
df2.head()
```

```python
labels=model2.labels_
centroids=model2.cluster_centers_
df2['labels']=labels
print(centroids)
```

```python
df2.drop(columns=['Cluster'],inplace=True)
df2
```

## Saving the result in the form of a csv file with clusters to which each person belong

```python
df2.to_csv('C:/Users/zhaleh/PhD_Projects/Divergent Paths of Beat Perception_3 Modalities/results/Three_Tactile_clusters.csv',index=None)
```

## Making necessary file for visualizing clusters with bars and individual points

```python
df_tact=pd.read_csv('C:/Users/zhaleh/PhD_Projects/Divergent Paths of Beat Perception_3 Modalities/results/Three_Tactile_clusters.csv')
cluster1tact=df_tact[df_tact['labels']==0].reset_index()
len(cluster1tact)
```

```python
cluster2tact=df_tact[df_tact['labels']==1].reset_index()
len(cluster2tact)
```

```python
cluster3tact=df_tact[df_tact['labels']==2].reset_index()
len(cluster3tact)
```

```python
########################################################## cluster 1
tactC1=pd.Series(np.arange(1,len(cluster1tact)*3+1))
for i in range(0,len(cluster1tact)*3):
    if i < len(cluster1tact):
        tactC1[i]='Single duration timing'
    elif i < len(cluster1tact)*2:
        tactC1[i]='Sequence timing'
    else:
        tactC1[i]='Beat-based timing'
dfC1=pd.concat([cluster1tact['TactSD'],cluster1tact['TactIrreg'],cluster1tact['TactReg']],axis=0,ignore_index=True)

dfCluster1=pd.DataFrame({'Timing.level':tactC1,'Cluster.1':dfC1})
########################################################## cluster 2
tactC2=pd.Series(np.arange(1,len(cluster2tact)*3+1))
for j in range(0,len(cluster2tact)*3):
    if j < len(cluster2tact):
        tactC2[j]='Single duration timing'
    elif j < len(cluster2tact)*2:
        tactC2[j]='Sequence timing'
    else:
        tactC2[j]='Beat-based timing'
dfC2=pd.concat([cluster2tact['TactSD'],cluster2tact['TactIrreg'],cluster2tact['TactReg']],axis=0,ignore_index=True)

dfCluster2=pd.DataFrame({'Timing.level2':tactC2,'Cluster.2':dfC2})
########################################################## cluster 3
tactC3=pd.Series(np.arange(1,len(cluster3tact)*3+1))
for j in range(0,len(cluster3tact)*3):
    if j < len(cluster3tact):
        tactC3[j]='Single duration timing'
    elif j < len(cluster3tact)*2:
        tactC3[j]='Sequence timing'
    else:
        tactC3[j]='Beat-based timing'
dfC3=pd.concat([cluster3tact['TactSD'],cluster3tact['TactIrreg'],cluster3tact['TactReg']],axis=0,ignore_index=True)

dfCluster3=pd.DataFrame({'Timing.level3':tactC3,'Cluster.3':dfC3})
######################################################### Creating the final file 
final=pd.concat([dfCluster1,dfCluster2,dfCluster3],axis=1)
final.to_csv('C:/Users/zhaleh/PhD_Projects/Divergent Paths of Beat Perception_3 Modalities/results/Three_Tactile_clusters_2.csv',index=None)
```
